**SEABORN **


Seaborn is a Python library built on top of Matplotlib. It makes beautiful, colorful, and easy-to-understand statistical plots with just a few lines of code

```python
import pandas as  pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("airlines_flights_data.csv")
```

```python
df
```

**LINE PLOT**

	Trends over time
	Connects points with lines (great for time-series data).

```python
sns.lineplot(data=df,x="airline",y="price",estimator="mean",hue="stops",size=50,style="stops",palette="Accent",markers=["*",">"],legend=False)
plt.title("Average price by airline")
plt.xlabel("Airlines")
plt.ylabel("Avg_price")
plt.grid()
plt.show()
```

```python
avg_price=df.groupby("airline")["price"].mean()
print(avg_price)
```

**BAR PLOT**

 Compare values,
	Shows bars for averages or totals (e.g., avg. salary per department).

```python
sns.barplot(data=df,x="stops",y="price",estimator="mean",hue="stops",hue_order=["zero","one","two_or_more"],orient="v",saturation=100)
plt.title("Bar chart for average price by class")
sns.set(style="darkgrid")
plt.xlabel("Class")
plt.ylabel("Average Price of Flights")
plt.grid()
plt.show()
```

**HISTOGRAM**

See distribution

Shows how often different values occur (e.g., age ranges).

```python
sns.set_style("white")
sns.histplot(df["price"],bins=[3000,6000,9000,12000,15000,18000,21000,24000,27000],color="red",kde=True)
plt.grid(False)
plt.show()
```

**SCATTER PLOT**

Compare 2 things

Puts dots on the chart to show relationships between two numeric columns.

```python
var=pd.DataFrame(df.head(500))
```

```python
m={"SpiceJet":"*","AirAsia":"o","Vistara":">","GO_FIRST":"<","Indigo":"^","Air_India":"*"}
sns.scatterplot(data=var,x="duration",y="price",hue="airline",style="airline",markers=m)
plt.show()
```

**HEATMAP PLOT**

Color-coded matrix

Shows data as a grid of colored boxes (used a lot in correlations).

```python
df.columns
```

```python
h=df[["days_left"]]
```

```python
sns.heatmap(h,vmin=0,vmax=40,cmap="gist_heat")
plt.show()
```

**COUNT PLOOT**

Count things

Counts how many times each category appears (e.g., number of people in each city).

```python
plt.figure(figsize=(8,4))
sns.countplot(data=df,x="airline",hue="stops",color="red",saturation=0.8)
plt.show()
```

**VIOLIN PLOT**

Spread + shape

Like a boxplot, but also shows how values are spread out (density).

```python
sns.violinplot(data=df,x="stops",y="price",hue="stops")
plt.show()
```

**PAIR PLOT**

Multi-relationships

Makes multiple scatterplots for all columns at once.

```python
pair=df[["duration","days_left","price"]]
```

```python
sns.pairplot(data=pair,diag_kind="hist")
plt.title("Compare kmer_3_freq, GC_Content, and base counts (Num_A, Num_T, Num_C, Num_G) between sequences with Mutation_Flag = 1 vs 0")
plt.show()
```

**STRIP PLOT**

Dots on line

Dots on a line to show individual points per category.

```python
sns.stripplot(data=df,x="airline",y="price",hue="airline",jitter=0.5)
plt.title("Strip plot to represent airline prices")
plt.xlabel("Airlines")
plt.ylabel("Prices")
plt.show()
```

**BOX PLOT**

See spread & outliers

Shows minimum, maximum, and middle values (good for spotting unusual values).

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10,6))
sns.boxplot(data=df,x="airline",y="days_left",hue="airline",showmeans=True,palette="plasma")
plt.suptitle("Box plot for days lef for the flight by airlines",fontsize=16,y=1.02)
plt.show()
```

**FACTOR PLOT** **Or CATPLOT **

A factor plot is a general-purpose plot in Seaborn used to visualize categorical data.

```python
sns.catplot(data=df,x="airline",y="price",kind="boxen",hue="airline") #here we use the kind function that helps to draw different type of the plots
plt.show()
#Options are 'strip', 'swarm', 'box', 'boxen', 'violin', 'bar', 'count', and 'point'.
```

**STYLING IN SEABORN**

```python
plt.figure(figsize=(10,6))
sns.set_style("darkgrid")
sns.set_context("poster",font_scale=1)
sns.barplot(data=df,x="airline",y="price")
sns.despine()
plt.show()
```

**Multiple plot ( Facet - Grid ) in seaborn**

```python
fg = sns.FacetGrid(df, height=6, aspect=2)
fg.map_dataframe(sns.barplot, x="airline", y="price")
plt.show()
```
